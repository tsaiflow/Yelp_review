```python
# train, test, validation split

le = LabelEncoder()
le.fit(df['Label'])
allClasses = le.transform(df['Label'])
allFeatures = df.drop(['Label'], axis=1)

X_trainAndTest, X_validation, y_trainAndTest, y_validation = train_test_split(allFeatures, allClasses, test_size=0.20, random_state=42)
X_train, X_test, y_train, y_test = train_test_split(X_trainAndTest, y_trainAndTest, test_size=0.20, random_state=42)
```

```python
# X_train
```

## 1.1 Number of Words

```python
X_train['word_count'] = X_train['text'].apply(lambda x: len(str(x).split(" ")))
X_train.head()
```

## 1.2 Number of Characters

```python
X_train['char_count'] = X_train['text'].str.len() ## this also includes spaces
X_train.head()
```

## 1.3 Average Word Length

```python
def avg_word(sentence):
  words = sentence.split()
  return (sum(len(word) for word in words)/len(words))
```

```python
X_train['avg_word'] = X_train['text'].apply(lambda x: avg_word(x))
X_train.head()
```

## 1.4 Number of Stopwords

```python
from nltk.corpus import stopwords
stop = stopwords.words('english')

X_train['stopwords'] = X_train['text'].apply(lambda x: len([x for x in x.split() if x in stop]))
X_train[['text','stopwords']].head()
```

## 1.5 Number of Special Characters

```python
X_train['hastags'] = X_train['text'].apply(lambda x: len([x for x in x.split() if x.startswith('#')]))
X_train[['text','hastags']].head()
```

## 1.6 Number of numerics

```python
X_train['numerics'] = X_train['text'].apply(lambda x: len([x for x in x.split() if x.isdigit()]))
X_train[['text','numerics']].head()
```

## 1.7 Number of Uppercase words

```python
X_train['upper'] = X_train['text'].apply(lambda x: len([x for x in x.split() if x.isupper()]))
X_train[['text','upper']].head()
```

# 2. Basic Pre-processing

## 2.1 Lower case

```python
X_train['text'] = X_train['text'].apply(lambda x: " ".join(x.lower() for x in x.split()))
X_train['text'].head()
```

## 2.2 Removing Punctuation

```python
X_train['text'] = X_train['text'].str.replace('[^\w\s]','')
X_train['text'].head()
```

## 2.3 Removal of Stop Words

```python
from nltk.corpus import stopwords
stop = stopwords.words('english')
X_train['text'] = X_train['text'].apply(lambda x: " ".join(x for x in x.split() if x not in stop))
X_train['text'].head()
```

## 2.4 Common word removal

```python
freq = pd.Series(' '.join(X_train['text']).split()).value_counts()[:10]
```

```python
freq = list(freq.index)
X_train['text'] = X_train['text'].apply(lambda x: " ".join(x for x in x.split() if x not in freq))
X_train['text'].head()
```

## 2.5 Rare words removal

```python
freq = pd.Series(' '.join(X_train['text']).split()).value_counts()[-10:]
```

```python
# freq
```

```python
freq = list(freq.index)
X_train['text'] = X_train['text'].apply(lambda x: " ".join(x for x in x.split() if x not in freq))
X_train['text'].head()
```

## 2.6 Spelling Correction

```python
from textblob import TextBlob
X_train['text'].apply(lambda x: str(TextBlob(x).correct()))
```

```python
X_train.iloc[1][0]
```

## 2.7 Tokenization

```python
TextBlob(X_train['text'][1]).words
```

## 2.8 Stemming

```python
from nltk.stem import PorterStemmer
st = PorterStemmer()
X_train['text'].apply(lambda x: " ".join([st.stem(word) for word in x.split()]))
```

```python
# len(X_train)
```

## 2.9 Lemmatization

```python
from textblob import Word
X_train['text'] = X_train['text'].apply(lambda x: " ".join([Word(word).lemmatize() for word in x.split()]))
X_train['text'].head(10)
```
